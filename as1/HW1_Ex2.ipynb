```python
import numpy as np                 # εισάγουμε τη βιβλιοθήκη numpy και θα τη "φωνάζουμε" np
import matplotlib.pyplot as plt    # εισάγουμε τη συνάρτηση pyplot της βιβλιοθήκης matplotlib και θα τη "φωνάζουμε" plt
```

Ο παρακάτω κώδικας εκτυπώνει τμήμα της γραφικής παράστασης του *μιγαδικού* σήματος
$$x(t)=\mathrm{Re}\{-(3+2j)e^{j2\pi 12t} \}$$    
παίρνοντας τιμές του σήματος $x(t)$ ανά $1/6000$ δευτερόλεπτα$^1$ στο διάστημα $[-0.05, 0.15]$ s. Δηλ. $6000$ τιμές της συνάρτησης $x(t)$ σε ένα δευτερόλεπτο!

Διαβάστε το πιο κάτω κελί, επιλέξτε το, και πατήστε Shift + Enter για να το τρέξετε.

```python
f0 = 12                                                    # μια τιμή για τη συχνότητα του εκθετικού
Z = -(3+2j)                                                # ο phasor του μιγαδικού εκθετικού
dt = 1/6000                                                # ο ρυθμός που παίρνουμε δείγματα από τη συνεχή συνάρτηση x(t)
t = np.arange(start=-0.05, stop=0.15, step=dt)             # ο άξονας του χρόνου: από -0.05 ως 0.15, με "βήμα" dt
xx = (Z * np.exp(1j*2*np.pi*f0*t)).real                    # η συνάρτησή μας!

plt.figure()                                               # Απεικόνιση: ανοίγουμε ένα figure
plt.plot(t, xx)                                            # Καλούμε την plt.plot για την απεικόνιση
plt.grid(linestyle=':')                                    # Ομορφαίνουμε 
plt.xlabel('Time (s)')                                     # Ομορφαίνουμε 
plt.title('Signal x(t)')                                   # Ομορφαίνουμε  
plt.xlim([min(t), max(t)])                                 # Ομορφαίνουμε 
plt.show()                                                 # Ομορφαίνουμε 
```

Διαβάστε λοιπόν προσεκτικά τον παραπάνω κώδικα και παραδώστε κώδικα που τυπώνει στο ίδιο διάστημα και με το ίδιο $dt$:

(α') το σήμα
$$x(t) = 7\cos\left(2\pi 20t + \frac{\pi}{2} \right) + 6\cos(2\pi 20t)$$

(β') το πραγματικό και το φανταστικό μέρος του σήματος
$$z(t) = 2.394e^{j(2\pi 80t+0.677)}$$

Παραδώστε τον κώδικα και τις γραφικές παραστάσεις που προκύπτουν.

```python
dt = 1/6000                                            # Σας δίνουμε το "βημα" ...
t = np.arange(start=-0.05, stop=0.15, step=dt)         # ...και τον άξονα του χρόνου t

# (a') 
xx = 7 * np.cos(2 * np.pi * 20 * t + np.pi/2) + 6 * np.cos(2 * np.pi * 20 * t)

plt.figure()
plt.plot(t, xx)
plt.grid(linestyle=':')
plt.xlabel('Time (s)')
plt.title('Signal')
plt.xlim([min(t), max(t)])
plt.show()
```

```python
# (b')
xx = 2.394 * np.e * 1j * (2 * np.pi * 80 * t + 0.677)

# Τυπώνουμε το πραγματικό μέρος...
plt.figure()
plt.plot(t, xx.real)
plt.grid(linestyle=':')
plt.xlabel('Time (s)')
plt.title('Real Part')
plt.xlim([min(t), max(t)])
plt.draw()

# ...και το φανταστικό μέρος
plt.figure()
plt.plot(t, xx.imag)
plt.grid(linestyle=':')
plt.xlabel('Time (s)')
plt.title('Imaginary Part')
plt.xlim([min(t), max(t)])
plt.draw()

plt.show()
```
