```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

Η απεικόνιση μιγαδικών συναρτήσεων είναι ελαφρά διαφορετική. Δεδομένου ότι οι συναρτήσεις αυτές είναι της μορφής $f : A \subset \mathbb{C} \rightarrow \mathbb{C}$, δεν μπορούμε να τις απεικονίσουμε ρητά - η συνάρτηση είναι τεσσάρων διαστάσεων, εν γένει. Θα απεικονίσουμε είτε το πραγματικό και το φανταστικό μέρος, είτε το μέτρο και τη φάση τους. Για παράδειγμα, έστω η μιγαδική συνάρτηση

$$ f(z)=3\cos(z)$$

την οποία θέλουμε να απεικονίσουμε στο μιγαδικό επίπεδο, στο "πλέγμα" $[-5,5] \times [-5, 5]$. 

Οι παρακάτω εντολές πραγματοποιούν την απεικόνιση του πραγματικού και του φανταστικού μέρους της συνάρτησης:

```python
range1 = np.arange(start=-5, stop=5, step=0.1)     # ο άξονας των πραγματικών (ο ίδιος θα είναι και ο φανταστικός, αφού το πλέγμα είναι συμμετρικό)
x,y = np.meshgrid(range1, range1, indexing='ij')   # δημιουργία πλέγματος
Z = x + 1j*y                                       # υπολογισμός του μιγαδικού Ζ στο πλέγμα
f = 3*np.cos(Z)                                    # υπολογισμός της συνάρτησης f(Z)
```

```python
plt.figure(figsize=(12,12))                           # Ορίζουμε αν θέλουμε το μέγεθος του γραφήματος

ax1 = plt.subplot(2, 2, 1, projection='3d')           # Δηλώνουμε ότι θα έχει δυο μέρη, σε 3Δ προβολή, και ότι τώρα θα ασχοληθούμε με το αριστερό
ax1.plot_surface(X=x, Y=y, Z=f.real, cmap=cm.viridis) # Τυπώνουμε το πραγματικό μέρος (το cmap αφορά το χρωματισμό)
plt.xlabel('$\mathfrak{Re}\{z\}$')                    # Ομορφαίνουμε
plt.ylabel('$\mathfrak{Im}\{z\}$')                    # --||--
plt.title('$\mathfrak{Re}\{3\cos(z)\}$')              # --||--

ax2 = plt.subplot(2, 2, 2, projection='3d')           # Δηλώνουμε ότι θα έχει δυο μέρη, σε 3Δ προβολή, και ότι τώρα θα ασχοληθούμε με το δεξί
ax2.plot_surface(X=x, Y=y, Z=f.imag, cmap=cm.viridis) # Τυπώνουμε το φανταστικό μέρος (το cmap αφορά το χρωματισμό)
plt.xlabel('$\mathfrak{Re}\{z\}$')                    # Ομορφαίνουμε
plt.ylabel('$\mathfrak{Im}\{z\}$')                    # --||--
plt.title('$\mathfrak{Im}\{3\cos(z)\}$')              # --||--
```

Χρησιμοποιείστε τις συναρτήσεις `abs`, `angle`, `exp` της `NumPy` για να απεικονίσετε το μέτρο $|f(z)|$ και τη φάση $\phi(z)$ της παραπάνω συνάρτησης στο ίδιο πεδίο ορισμού. 

Επιπλέον, σχεδιάστε το πραγματικό και το φανταστικό μέρος της συνάρτησης $$f(z)=e^z$$ στο "πλέγμα" $[-5,5]\times[-10,10]$.

```python
range1 = np.arange(start=-5, stop=5, step=0.1)   # Σας δίνουμε τον πρώτο άξονα, [-5,5]
range2 = np.arange(start=-10, stop=10, step=0.1) # Φτιάξτε το δεύτερο άξονα, [-10,10]
x,y = np.meshgrid(range1, range2, indexing='ij') # δημιουργία πλέγματος
Z = x + 1j*y                                     # υπολογισμός μιγαδικού Ζ στο πλέγμα
f = np.e ** Z                                    # υπολογισμός συνάρτησης
```

```python
plt.figure(figsize=(10,10))                       # Δημιουργία γραφήματος (χρησιμοποιήστε τον έτοιμο κώδικα παραπάνω, με κατάλληλη τροποποίηση)

# Insert code here
ax1 = plt.subplot(2, 2, 1, projection='3d')           # Δηλώνουμε ότι θα έχει δυο μέρη, σε 3Δ προβολή, και ότι τώρα θα ασχοληθούμε με το αριστερό
ax1.plot_surface(X=x, Y=y, Z=f.real, cmap=cm.viridis) # Τυπώνουμε το πραγματικό μέρος (το cmap αφορά το χρωματισμό)
plt.xlabel('$\mathfrak{Re}\{z\}$')                    # Ομορφαίνουμε
plt.ylabel('$\mathfrak{Im}\{z\}$')                    # --||--
plt.title('$\mathfrak{Re}\{3^Z\}$')              # --||--

ax2 = plt.subplot(2, 2, 2, projection='3d')           # Δηλώνουμε ότι θα έχει δυο μέρη, σε 3Δ προβολή, και ότι τώρα θα ασχοληθούμε με το δεξί
ax2.plot_surface(X=x, Y=y, Z=f.imag, cmap=cm.viridis) # Τυπώνουμε το φανταστικό μέρος (το cmap αφορά το χρωματισμό)
plt.xlabel('$\mathfrak{Re}\{z\}$')                    # Ομορφαίνουμε
plt.ylabel('$\mathfrak{Im}\{z\}$')                    # --||--
plt.title('$\mathfrak{Im}\{3^Z\}$')              # --||--
# End of code
```
