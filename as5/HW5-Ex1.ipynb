```python
from scipy.signal import residue
```

Έχετε ήδη καταλάβει ότι το ανάπτυγμα σε μερικά κλάσματα είναι μια σωτήρια 😃 τεχνική όσον αφορά αντίστροφους μετασχηματισμούς, τόσο Fourier όσο και Laplace. Φυσικά, η Python δε θα μπορούσε να μην έχει την αντίστοιχη συνάρτηση για να μας διευκολύνει, και αυτή είναι η $\texttt{residue}$ της $\texttt{SciPy}$ βιβλιοθήκης. Έστω λοιπόν ότι έχουμε μια ρητή συνάρτηση του $j2\pi f$ ή του $s$. Η σύνταξή της εντολής για να μας κάνει το ανάπτυγμα είναι η ακόλουθη:

$$\textrm{r, p, k = residue(b, a)}$$

με 
1. $b$ ένα διάνυσμα που περιέχει τους συντελεστές του πολυωνύμου (του $s$ ή $j2\pi f$) του αριθμητή σε φθίνουσα σειρά βαθμού (πρώτα ο συντελεστής του όρου με το μεγαλύτερο βαθμό, κλπ, κλπ, και τελευταία η σταθερά), και 
2. $a$ ένα διάνυσμα που περιέχει τους συντελεστές του πολυωνύμου (του $s$ ή $j2\pi f$) του παρονομαστή, επίσης σε φθίνουσα σειρά βαθμού (πρώτα ο συντελεστής του όρου με το μεγαλύτερο βαθμό, κλπ, κλπ, και τελευταία η σταθερά).

Η συνάρτηση επιστρέφει τρία διανύσματα: 
1. το διάνυσμα $r$ περιέχει τους αριθμητές των μερικών κλασμάτων που προκύπτουν, 
2. το διάνυσμα $p$ περιέχει τους πόλους των απλών κλασμάτων, δηλ. τις ρίζες των παρονομαστών των μερικών κλασμάτων, και 
3. το διάνυσμα $k$ περιέχει συντελεστές όρων που προκύπτουν κατά τη διαίρεση πολυωνύμων στην περίπτωση που η ρητή συνάρτηση έχει βαθμό πολυωνύμου αριθμητή μεγαλύτερο ή ίσο με τον αντίστοιχο του παρονομαστή.


Ας δούμε ένα παράδειγμα. Έστω:

$$X(s) = \frac{s^2 + 2s + 1}{s^3 -3s +2}$$

Ο αριθμητής έχει συντελεστές $[1, 2, 1]$ και ο παρονομαστής $[1, 0, -3, 2]$, καθώς ο δευτεροβάθμιος όρος δεν υπάρχει στο πολυώνυμο του τελευταίου, κι έτσι ο συντελεστής του είναι $0$. 

Εκτελούμε λοιπόν:

```python
b = [1, 2, 1]
a = [1, 0, -3, 2]

r, p, k = residue(b, a)

print(f"r = {r} \np = {p} \nk = {k}")
```

Από αυτό καταλαβαίνουμε αμέσως ότι επειδή το διάνυσμα $k$ είναι άδειο, δεν υπάρχουν όροι από τη διαίρεση των πολυωνύμων - λογικό κι αναμενόμενο, καθώς ο αριθμητής μας είναι δευτεροβάθμιος και ο παρονομαστής τριτοβάθμιος. Από τα υπόλοιπα, παρατηρούμε ότι μπορούμε να γράψουμε το αποτέλεσμα του αναπτύγματος ως

### $$X(s) = \frac{s^2 + 2s + 1}{s^3 - 3s + 2} = \frac{0.\bar{1}}{s+2} +  \frac{1.\bar{3}}{s-1} + \frac{0.\bar{8}}{s-1}  = \frac{\frac{1}{9}}{s+2} + \frac{\frac{4}{3}}{s-1} + \frac{\frac{8}{9}}{s-1}$$

Δυστυχώς, η Python μας αφήνει εδώ. 😃 Στη συνέχεια, ανάλογα με το πεδίο σύγκλισης, μπορούμε εμείς να γυρίσουμε τα σήματά μας πίσω στο χρόνο μέσω των γνωστών πινάκων μετασχηματισμού Laplace.
